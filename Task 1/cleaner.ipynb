````python
import pandas as pd
````

````python
excel_file = './KPMG_VI_New_raw_data_update_final.xlsx'

sheets = pd.read_excel(excel_file, sheet_name=None)
dataframes = []
for sheet_name, sheet_data in sheets.items():
    df = pd.DataFrame(sheet_data)
    dataframes.append(df)
````

# We will take the first file from the data Transaction dataset

````python
df = dataframes[1]
````

Now we will have to remove teh default header as it is invalid and make the first row as the header

````python
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header

df.head()
````

````python
df['transaction_date'] = pd.to_datetime(df['transaction_date']).dt.date
````

````python
df.info()
````

````python
df.duplicated().sum()
````

````python
df.isna().sum()
````

```
Now we are going to either fill the values with the appropriate value or remove the data in order to get a dataset having no missing value
```

````python
columns_to_fill = ['online_order', 'brand',
                   'product_line', 'product_class', 'product_size', 'product_first_sold_date']
df[columns_to_fill] = df[columns_to_fill].fillna(method='ffill')
````

````python
df.dtypes
````

```
Now we are going to change the Dtypes of some of the columns to their respective types accourding to the type of data in them
```

````python
columns_to_convert = ['transaction_id', 'product_id',
                      'customer_id', 'product_first_sold_date']

for column in columns_to_convert:
    df[column] = df[column].astype(int)
````

````python
columns_to_convert = ['list_price', 'standard_cost']

for column in columns_to_convert:
    df[column] = df[column].astype(float)
````

````python
df.head()
````

````python
df['standard_cost']=df['standard_cost'].fillna(df['standard_cost'].mean())
````

````python
df.isna().sum()
````

Now that we have cleaned the data we will save it in the original data

````python
dataframes[1]=df
````
